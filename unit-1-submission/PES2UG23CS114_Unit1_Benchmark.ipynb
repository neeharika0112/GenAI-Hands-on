## Objective
To evaluate architectural differences between BERT, RoBERTa, and BART by forcing them to perform tasks they are not primarily designed for, and observing the outcomes.

```python
from transformers import pipeline
```

## Models Used
- BERT: bert-base-uncased (Encoder-only)
- RoBERTa: roberta-base (Optimized Encoder-only)
- BART: facebook/bart-base (Encoder–Decoder)


Experiment 1: Text Generation
Task: Try to generate text using the prompt: "The future of Artificial Intelligence is"

Code Hint: pipeline('text-generation', model='...')
Hypothesis: Which models will fail? Why? (Hint: Can an Encoder generate new tokens easily?)

```python
models = {
    "BERT": "bert-base-uncased",
    "RoBERTa": "roberta-base",
    "BART": "facebook/bart-base"
}

prompt = "The future of Artificial Intelligence is"

results_gen = {}

for name, model in models.items():
    try:
        generator = pipeline("text-generation", model=model)
        output = generator(prompt, max_length=30)
        results_gen[name] = output
        print(f"\n{name} Output:\n", output)
    except Exception as e:
        results_gen[name] = str(e)
        print(f"\n{name} Error:\n", e)
```

Experiment 2: Masked Language Modeling (Missing Word)
Task: Predict the missing word in: "The goal of Generative AI is to [MASK] new content."

Code Hint: pipeline('fill-mask', model='...')
Hypothesis: This is how BERT/RoBERTa were trained. They should perform well.

```python
sentence = "The goal of Generative AI is to [MASK] new content."

results_mask = {}

for name, model in models.items():
    try:
        fill_mask = pipeline("fill-mask", model=model)
        output = fill_mask(sentence)
        results_mask[name] = output
        print(f"\n{name} Predictions:")
        for pred in output[:3]:
            print(pred["token_str"], "-", pred["score"])
    except Exception as e:
        results_mask[name] = str(e)
        print(f"\n{name} Error:\n", e)
```

Experiment 3: Question Answering
Task: Answer the question "What are the risks?" based on the context: "Generative AI poses significant risks such as hallucinations, bias, and deepfakes."

Code Hint: pipeline('question-answering', model='...')
Note: Using a "base" model (not fine-tuned for SQuAD) might yield random or poor results. Observe this behavior.

```python
qa_input = {
    "context": "Generative AI poses significant risks such as hallucinations, bias, and deepfakes.",
    "question": "What are the risks?"
}

results_qa = {}

for name, model in models.items():
    try:
        qa = pipeline("question-answering", model=model)
        output = qa(qa_input)
        results_qa[name] = output
        print(f"\n{name} Answer:\n", output)
    except Exception as e:
        results_qa[name] = str(e)
        print(f"\n{name} Error:\n", e)
```

## Observation Table

| Task | Model | Classification | Observation | Architectural Reason |
|----|----|----|----|----|
| Generation | BERT | Failure | Could not generate text / threw error | Encoder-only models are not trained for next-token generation |
| Generation | RoBERTa | Failure | Similar failure as BERT | Optimized encoder, still no decoding head |
| Generation | BART | Success | Generated coherent continuation | Encoder–decoder architecture supports text generation |
| Fill-Mask | BERT | Success | Predicted words like "generate", "create" | Trained using Masked Language Modeling |
| Fill-Mask | RoBERTa | Success | High-confidence predictions | Improved MLM training over BERT |
| Fill-Mask | BART | Partial Success | Predictions less confident | Not primarily trained for MLM |
| QA | BERT | Partial Success | Returned short or vague answers | Not fine-tuned for QA |
| QA | RoBERTa | Partial Success | Slightly better phrasing | Stronger encoder representations |
| QA | BART | Partial Success | Sometimes irrelevant answer | Base model, not QA fine-tuned |


## Understanding & Learnings

### What I Understood
- Encoder-only models like BERT and RoBERTa are designed for understanding tasks, not generation.
- Encoder–decoder models like BART can perform both understanding and generation tasks.
- Model architecture strongly influences task suitability.
- Using base models without fine-tuning leads to poor results in specialized tasks like QA.

### What I Built
- Benchmarked three transformer models across generation, fill-mask, and QA tasks.
- Observed success and failure cases to understand architectural limitations.
