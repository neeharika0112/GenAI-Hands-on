This project implements a Question Answering (QA) system using the
`distilbert-base-cased-distilled-squad` model from Hugging Face.

The system takes a context paragraph and a question as input and extracts
the most relevant answer directly from the context.

## Why DistilBERT?
- Faster than BERT
- Lightweight
- Optimized for extractive Question Answering
- Suitable for real-time applications like FAQ bots


```python
!pip install transformers torch
```

```python
from transformers import pipeline
```

```python
qa_pipeline = pipeline(
    "question-answering",
    model="distilbert-base-cased-distilled-squad"
)

print("Model loaded successfully!")
```

```python
context = """
Generative AI refers to artificial intelligence systems that can generate
new content such as text, images, audio, and video. While these systems
offer powerful capabilities, they also pose risks including hallucinations,
bias, misinformation, and deepfake generation. Responsible AI practices
are required to mitigate these risks.
"""
```

```python
question = "What are the risks of Generative AI?"
```

```python
result = qa_pipeline(
    question=question,
    context=context
)

result
```

```python
print("Question:", question)
print("Answer:", result["answer"])
print("Confidence Score:", result["score"])

```

```python
question2 = "Why are responsible AI practices needed?"

result2 = qa_pipeline(
    question=question2,
    context=context
)

print("Question:", question2)
print("Answer:", result2["answer"])
print("Confidence Score:", result2["score"])
```

## Observations

- The DistilBERT model accurately extracts answers directly from the context.
- The answers are concise and relevant.
- The confidence score indicates how certain the model is.
- Since this is extractive QA, the model does not generate new text.


## Limitations

- The model cannot answer questions outside the given context.
- It cannot generate explanations or summaries.
- Performance depends heavily on the quality of the context.


## Conclusion

This project demonstrates how DistilBERT can be used to build a fast and
accurate extractive Question Answering system. Due to its lightweight
architecture, it is suitable for real-time applications such as FAQ bots,
educational assistants, and customer support systems.
