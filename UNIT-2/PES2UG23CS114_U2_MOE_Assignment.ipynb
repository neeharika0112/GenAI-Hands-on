```python
MODEL_CONFIG = {
    "technical": "Handles coding/debugging issues",
    "billing": "Handles payments/refunds",
    "general": "Handles general queries"
}
```

```python
def route_prompt(user_input):
    user_input = user_input.lower()

    if "error" in user_input or "code" in user_input or "bug" in user_input:
        return "technical"
    elif "charged" in user_input or "payment" in user_input or "refund" in user_input:
        return "billing"
    else:
        return "general"
```

```python
def process_request(user_input):
    category = route_prompt(user_input)

    print(f"🔀 Routed to: {category.upper()} expert\n")

    if category == "technical":
        return "This seems like a coding issue. Please check your logic or indexing."

    elif category == "billing":
        return "We’re sorry for the inconvenience. Please contact support for refund processing."

    else:
        return "Hello! How can I assist you today?"
```

```python
print(process_request("My Python code has an IndexError"))
print(process_request("I was charged twice"))
print(process_request("Tell me a joke"))
```

```python
def tool_expert(user_input):
    user_input = user_input.lower()

    if "bitcoin" in user_input:
        return "Current Bitcoin price is approximately $60,000 (mock data)."

    return None
```

```python
def process_request(user_input):

    # Tool interception
    tool_result = tool_expert(user_input)
    if tool_result:
        print(" Routed to: TOOL EXPERT\n")
        return tool_result

    # Normal routing
    category = route_prompt(user_input)

    print(f"🔀 Routed to: {category.upper()} expert\n")

    if category == "technical":
        return "This seems like a coding issue. Please check your logic or indexing."

    elif category == "billing":
        return "We’re sorry for the inconvenience. Please contact support for refund processing."

    else:
        return "Hello! How can I assist you today?"
```

```python
print(process_request("What is the current price of Bitcoin?"))
```
